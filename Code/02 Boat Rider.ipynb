```python
import csv

# Define the order table as a string (simulating a file)
order_data = """Order_ID\ttype\tFrom\tTo\tน้ำหนัก\tstart time\tarrived due date
o1\t\u0e02\u0e32\u0e02\u0e35\u0e49\u0e19\tcarrier1\tcustomer1\t10000\t12/1/2025\t15/1/2025
o2\t\u0e02\u0e32\u0e02\u0e35\u0e49\u0e19\tcarrier1\tcustomer2\t5000\t12/1/2025\t16/1/2025
o3\t\u0e02\u0e32\u0e02\u0e35\u0e49\u0e19\tcarrier2\tcustomer3\t10000\t12/1/2025\t16/1/2025"""

# Read the data as dictionary
orders = []
reader = csv.DictReader(order_data.splitlines(), delimiter='\t')
for row in reader:
    orders.append(row)

# Print the result
for order in orders:
    print(order)

from datetime import datetime

# Original data
orders = [
    {'Order_ID': 'o1', 'type': 'ขาขี้น', 'From': 'carrier1', 'To': 'customer1', 'น้ำหนัก': '10000', 'start time': '12/1/2025', 'arrived due date': '15/1/2025'},
    {'Order_ID': 'o2', 'type': 'ขาขี้น', 'From': 'carrier1', 'To': 'customer2', 'น้ำหนัก': '5000', 'start time': '12/1/2025', 'arrived due date': '16/1/2025'},
    {'Order_ID': 'o3', 'type': 'ขาขี้น', 'From': 'carrier2', 'To': 'customer3', 'น้ำหนัก': '10000', 'start time': '12/1/2025', 'arrived due date': '16/1/2025'}
]

# Mapping for translation
translations = {
    "ขาขี้น": "Import",
    "น้ำหนัก": "Weight"
}

# Convert data
converted_orders = []
for order in orders:
    converted_order = {
        "Order_ID": order["Order_ID"],
        "Type": translations.get(order["type"], order["type"]),
        "From": order["From"],
        "To": order["To"],
        "Weight": int(order["น้ำหนัก"]),
        "Start Time": datetime.strptime(order["start time"], "%d/%m/%Y"),
        "Arrived Due Date": datetime.strptime(order["arrived due date"], "%d/%m/%Y")
    }
    converted_orders.append(converted_order)

# Print converted data
for order in converted_orders:
    print(order)
```
